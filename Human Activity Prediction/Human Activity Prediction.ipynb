```python
import pandas as pd

df=pd.read_csv('train.csv')

df.head()
```

```python
df.shape
```

```python
features_train=df
features_train=df.drop('Activity',1)
```

```python
features_train.head()
```

```python
labels_train=df['Activity']
```

```python
test=pd.read_csv('test.csv')

features_test=test.drop('Activity',1)

labels_test=test['Activity']
```

```python
from sklearn.svm import SVC

clf=SVC()

clf.fit(features_train,labels_train)

prediction=clf.predict(features_test)

from sklearn.metrics import accuracy_score

acc=accuracy_score(prediction,labels_test)
```
